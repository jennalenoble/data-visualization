# Module 2: Visual Encoding Channels

### Area Charts

```python
import altair as alt
from vega_datasets import data
import pandas as pd
```

```python
gm = pd.read_csv("data/world-data-gapminder.csv", parse_dates = ['year'])
gm
```

```python
gm.info()
```

```python
alt.Chart('https://website.com/data.csv').mark_line().encode(
    x='column1:T', #Temporal
    y='column2:Q') #Quantitative
```

```python
gm_url = 'https://raw.githubusercontent.com/UBC-MDS/exploratory-data-viz/main/chapters/en/slides/module2/data/world-data-gapminder.csv'
alt.Chart(gm_url).mark_line().encode(
    x='year:T',
    y='population:Q')
```

```python
alt.Chart(gm_url).mark_line().encode(
    x='year:T',
    y='sum(population):Q')
```

```python
alt.Chart(gm_url).mark_line().encode(
    x='year:T',
    y=alt.Y('sum(population):Q', title='World population'))
```

```python
alt.Chart(gm_url).mark_area().encode(
    x='year:T',
    y=alt.Y('sum(population):Q', title='World population'))
```

```python
alt.Chart(gm_url).mark_area().encode(
    alt.X('year:T'),
    alt.Y('sum(population):Q', title='Cumulative population of stacked regions'),
    color='region:N')
```

```python
alt.Chart(gm_url).mark_line().encode(
    alt.X('year:T'),
    alt.Y('sum(population):Q', title='Cumulative population of stacked regions'),
    color='region:N')
```

```python
alt.Chart(gm_url).mark_area().encode(
    alt.X('year:T'),
    alt.Y('mean(children_per_woman):Q', title='Average children per woman worldwide'))
```

### Bar Charts and Histograms

```python
gm2018 = gm[gm['year']=='2018']
gm2018
```

```python
alt.Chart(gm2018).mark_bar().encode(
    alt.X('region'),
    alt.Y('sum(population)', title='Population'))
```

```python
alt.Chart(gm2018).mark_bar().encode(
    alt.X('sum(population)', title='Population'),
    alt.Y('region'))
```

```python
alt.Chart(gm2018).mark_bar().encode(
    alt.X('count()', title='Number of countries'),
    alt.Y('region'))
```

```python
alt.Chart(gm2018).mark_bar().encode(
    alt.X('count()', title='Number of countries'),
    alt.Y('region', sort='x'))
```

```python
alt.Chart(gm2018).mark_bar().encode(
    alt.X('count()', title='Number of countries'),
    alt.Y('region', sort=['Oceania', 'Americas', 'Europe', 'Asia', 'Africa']))
```

```python
alt.Chart(gm2018).mark_bar().encode(
    alt.X('count()', title='Number of countries'),
    alt.Y('region', sort='-x'))
```

```python
my_order = ['Africa', 'Europe', 'Oceania', 'Asia', 'Americas']
alt.Chart(gm2018).mark_bar().encode(
    alt.X('count()', title = "Count"),
    alt.Y('region', sort=my_order))
```

```python
alt.Chart(gm2018).mark_bar().encode(
    alt.X('life_expectancy'),
    alt.Y('count()'))
```

```python
alt.Chart(gm2018).mark_bar().encode(
    alt.X('life_expectancy', bin=True),
    alt.Y('count()'))
```

```python
alt.Chart(gm2018).mark_bar().encode(
    alt.X('life_expectancy', bin=alt.Bin(maxbins=30), title='Life expectancy'),
    alt.Y('count()'))
```

### Creating Subplots via Faceting

```python
alt.Chart(gm2018).mark_bar().encode(
    alt.X('life_expectancy', bin=alt.Bin(maxbins=30)),
    alt.Y('count()'),
    alt.Color('region'))
```

```python
alt.Chart(gm2018).mark_bar(opacity=0.7).encode(
    alt.X('life_expectancy', bin=alt.Bin(maxbins=30)),
    alt.Y('count()', stack=False),
    alt.Color('region'))
```

```python
alt.Chart(gm2018).mark_bar().encode(
    alt.X('life_expectancy', bin=alt.Bin(maxbins=30)),
    alt.Y('count()'),
    alt.Color('region')
).properties(width=200, height=150
            ).facet('region')
```

```python
alt.Chart(gm2018).mark_bar().encode(
    alt.X('life_expectancy', bin=alt.Bin(maxbins=30)),
    alt.Y('count()'),
    alt.Color('region')
).properties(width=200, height=50
            ).facet('region', columns=2)
```

```python
alt.Chart(gm2018).mark_bar().encode(
    alt.X('life_expectancy', bin=alt.Bin(maxbins=30)),
    alt.Y('count()'),
    alt.Color('region')
).properties(width=200, height=140
            ).facet('region', columns=3)
```

```python
alt.Chart(gm2018[gm2018['income_group'].isin(['High','Low'])]).mark_bar().encode(
    alt.X('life_expectancy', bin=alt.Bin(maxbins=30)),
    alt.Y('count()'),
    alt.Color('region')
).properties(width=200, height=140
            ).facet(column='region', row='income_group')
```

```python
income_order = ['High', 'Upper middle', 'Lower middle', 'Low']
alt.Chart(gm2018[gm2018['region'].isin(['Africa'])]).mark_point().encode(
    alt.X('children_per_woman'),
    alt.Y('child_mortality'),
    alt.Color('income_group', sort=income_order)
).properties(width=200, height=140
            ).facet('sub_region')
```

```python
alt.Chart(gm2018[gm2018['region'].isin(['Africa'])]).mark_bar().encode(
    alt.X('count()', title='Number of countries'),
    alt.Y('sub_region'),
    alt.Color('income_group', sort=income_order)
).properties(width=200)
```

```python
alt.Chart(gm2018[gm2018['region'].isin(['Africa'])]).mark_bar().encode(
    alt.X('count()', title='Number of countries'),
    alt.Y('income_group', sort=income_order),
    alt.Color('income_group', sort=income_order, legend=None)
).properties(width=200).facet('sub_region')
```
